```r
install.packages("ggplot2")
#Line, bar, histogram, Pie,stacked bar, scatter,box plot, grouped bar, word cloud.
attach(iris)
iris_demo <-iris
#View(iris_demo)
```

```r
str(iris_demo) #strucutre of the dataset
#?plot
plot(iris_demo$Sepal.Length,iris_demo$Species)

```

```r
plot(iris_demo$Sepal.Width,iris_demo$Species)
```

```r
plot(iris_demo$Petal.Length,iris_demo$Species)
```

```r
plot(iris_demo$Sepal.Length,iris_demo$Sepal.Width)
points(iris_demo$Petal.Length,iris_demo$Petal.Width,col="Red")
```

```r
plot(iris_demo$Sepal.Length,iris_demo$Sepal.Width,col=iris_demo$Species)
plot(iris_demo$Petal.Length,iris_demo$Petal.Width,col=iris_demo$Species,main="Dataset - Iris",sub="Petal",xlab="Length",ylab="Width")
legend(2,2,legend=c("S","V","C"),col=c("black","red","green"),pch=.5)
```

```r
#?hist
hist(iris_demo$Sepal.Width,main="Hist of sepal width",xlab="width",border="blue",col="green",breaks = c(2,7,9))

```

```r
hist(iris_demo$Sepal.Width,main="Hist of sepal width",xlab="width",border="blue",col="green",breaks = 1:6)
```

```r
hist(iris_demo$Sepal.Width,main="Hist of sepal width",xlab="width",border="blue",col="green",freq=F)
```

```r
# create histograms for each attribute
data(iris)
par(mfrow=c(1,4))
for(i in 1:4) {
  hist(iris[,i], main=names(iris)[i])
}


```

```r
par(mfrow=c(1,4))
for(i in 1:4) {
  #hist(iris[,i], main=names(iris)[i])
  boxplot(iris[,i],main=names(iris)[i])
  
}
```

```r
par(mfrow=c(1,1))
boxplot(iris_demo,las=2)
```

```r
#grab the outlier
boxplot(iris_demo$Sepal.Width)
outlier <- boxplot(iris_demo$Sepal.Width)
outlier
```

```r

#Notch
boxplot(iris_demo$Petal.Length~iris_demo$Species,col=c("red","green","purple"),xlab="Species",ylab="Petal length",notch=T)
```

```r
boxplot(iris_demo$Petal.Width~iris_demo$Species,col=c("red","green","purple"),xlab="Species",ylab="Petal width",notch=T)
```
