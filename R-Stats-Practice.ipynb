## Hypothesis testing

## 2 sample testing - 1 sided

#### Example 1

```r
girls = c(7,8,9,8,6,7,8,9,5,6,9,8)
boys = c(1,8,1,2,4,7,8,6,5,3,4,9)

## Ha: mean (girs) >> mean (boys)
## h0: mean(girls) <= mean(boys)

mean(girls)
mean(boys)
```

```r
t.test(girls,boys, alternative = "greater")
# if p-value is less then 0.05 then we will go with alternate hypothesis
# this we means 0.05 chances of making 5% perent type 1 error
```

```r
# From above as p-value is less then 0..05 - we are accepting alternate hypothesis - i.e, girls are better
# From abobe as confidence internal has no 0 (it is ranging from 1.10 to inf) - we can go with alternate hypothesis
```

```r
t.test(boys,girls, alternative = "greater")
```

```r
t.test(boys,girls, alternative = "greater")
```

#### Example 2

```r
girls1 = c(7,8,9,8,6,7,8,9,5,6,9,8)
boys1 = c(7,8,9,8,6,5,8,5,6,6,9,8)

## Ha: mean (girs) >> mean (boys)
## h0: mean(girls) <= mean(boys)

mean(girls1)
mean(boys1)
```

```r
t.test(girls1,boys1, alternative = "greater")
```

```r
# In above case shows 20% chance of making type 1 error
# -.55 - inf - this means difference in means can be anywhere between -.55 to inf - as there are values below 0
# - it means in some samples boys and girls means are same - means reject alternate hypothesis 
```
