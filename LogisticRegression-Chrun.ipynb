## We will go with day charges, even charges, voice mail plan, customer service calls, intl plans, account.length,

```r
df$VMail.Plan = as.factor(df$VMail.Plan)
df$Intl.Plan = as.factor(df$Intl.Plan)
table(df$VMail.Plan)
```

## Divide dataset into training and test set

```r
set.seed(7)
ids = sample(nrow(df),nrow(df)*0.8)
train = df[ids,]
test=df[-ids,]
```

```r
churn_model=glm(Churn ~ Account.Length + VMail.Plan +  Intl.Plan + CustServ.Calls +Day.Charge +Eve.Charge , data = train, family="binomial")
```

```r
names(df)
```

```r
summary(churn_model)
```

### Predicted test observations

```r
test$pred = predict(churn_model,newdata=test,type="response")
head(test$pred)
```

```r
test$pred_class = ifelse(test$pred>=0.5,1,0)
```

## Confusion matrix

```r
table(test$pred_class,test$Churn)
```
