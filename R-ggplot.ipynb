### Convert to factors
- Conceptually, factors are variables in R which take on a limited number of different values; such variables are often refered to as categorical variables. One of the most important uses of factors is in statistical modeling; since categorical variables enter into statistical models differently than continuous variables, storing data as factors insures that the modeling functions will treat such data correctly.
Factors in R are stored as a vector of integer values with a corresponding set of character values to use when the factor is displayed. The factor function is used to create a factor. The only required argument to factor is a vector of values which will be returned as a vector of factor values. Both numeric and character variables can be made into factors, but a factor's levels will always be character values. You can see the possible levels for a factor through the levels command. 

```r
## Note cylinder should be factors, compare output now with previous one - 5 and 7 are gone
ggplot(data=mtcars, aes(x=factor(cyl),y=mpg)) + geom_point()
```

### Scatter plots

```r
ggplot(data=mtcars,aes(x=factor(cyl),y=mpg)) + geom_point(shape=2,size=5)
```

### Use size in AES - helps in analyzing  3 parameters at one time

```r
## Use size in Aes
ggplot(data=mtcars,aes(x=factor(cyl),y=mpg,size=hp)) + geom_point(shape=2) #Now size is displayed based on horse power
```

#### Analyize 4 parameters at one time

```r
ggplot(data=mtcars,aes(x=wt,y=mpg,col=hp,size=disp)) + geom_point()
```

```r
ggplot(data=mtcars,aes(x=wt,y=mpg,col=hp,size=disp)) + geom_text(aes(label=cyl))
```

### Displaying text labels

```r
ggplot(data=mtcars,aes(x=wt,y=mpg)) + geom_text(aes(label=cyl))
#Whenever you want to introduce variablity use aes parameter
```

### Using more then one geom_

```r
ggplot(data=mtcars,aes(x=wt,y=mpg)) + geom_point() + geom_line(aes(y=qsec)) #techinal this graph doesn't make any sense - this is just for info
```
