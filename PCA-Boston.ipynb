### Issue with PCA - there is no straight forward way to tell the impact of output on single variable (for example we cannot tell '1 percent increase in crime rate than how much will medv value change??' - this cannot be answered using PCA)

### Hence PCA is mainly used to build predcitive model and is not used to build explanatory model
