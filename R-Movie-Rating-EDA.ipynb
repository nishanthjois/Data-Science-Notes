# Merge all 3 files

```r
?merge
```

```r
my_ratings = merge(ratings,movies,by="movie_id")
my_ratings
```

```r
my_user_ratings = merge(my_ratings,users,by="user_id")
my_user_ratings
```

# Check which movie has highest number of ratings

```r
# Add a column and name it dummy variable
# Assign value = 1 too all rows of dummy column
my_user_ratings$dummy = 1

# Count by movie id
highest_rating_movie_id = aggregate(my_user_ratings$dummy,by=list(unique.values=my_user_ratings$movie_id),FUN=sum)

# or Count by title
highest_rating_title = aggregate(my_user_ratings$dummy,by=list(unique.values=my_user_ratings$title),FUN=sum)
highest_rating_title
```

```r
# Order by decreasing order
highest_rating_title_sorted = highest_rating_title[order(highest_rating_title$x,decreasing = TRUE),]
highest_rating_title_sorted
```

# Mean rating of each movie

```r
mean_rating_title = aggregate(my_user_ratings$dummy,by=list(unique.values=my_user_ratings$title),FUN=mean)
```
